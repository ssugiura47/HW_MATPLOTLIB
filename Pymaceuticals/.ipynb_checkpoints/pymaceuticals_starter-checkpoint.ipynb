```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
len(merge_df["Mouse ID"].value_counts())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupemouseID = merge_df.loc[merge_df.duplicated(subset=['Mouse ID','Timepoint']),'Mouse ID'].unique()
dupemouseID = dupemouseID[0]
dupemouseID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dupemouse = merge_df.loc[merge_df["Mouse ID"] == dupemouseID,:]
dupemouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
dupemouse = merge_df[merge_df['Mouse ID'] == dupemouseID].index
merge_df.drop(dupemouse, inplace=True)
merge_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(merge_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean = merge_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median = merge_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
variance = merge_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_dev = merge_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()

regimensummary = np.array(mean.index.tolist())
regimen_mean = np.array(mean.values.tolist())
regimen_median = np.array(median.values.tolist())
regimen_var = np.array(variance.values.tolist())
regimen_std_dev = np.array(std_dev.values.tolist())

Summary_df = pd.DataFrame({
    "Regimen" : regimensummary,
    "Mean": regimen_mean,
    "Median": regimen_median,
    "Variance": regimen_var,
    "Standard Deviation": regimen_std_dev,

})
Summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
regimen_df = merge_df.groupby("Drug Regimen").describe()
regimen_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

RegimenTimepoint_df = merge_df.groupby("Mouse ID")["Timepoint"].max().reset_index()
mergeRegimenTimepoint_df = pd.merge(merge_df, RegimenTimepoint_df)
mergeRegimenTimepoint_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)

fourRegimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

FinalTumorVol = []

for regimen in fourRegimen:
    final_tumor_vol = mergeRegimenTimepoint_df.loc[mergeRegimenTimepoint_df["Drug Regimen"] == regimen, 'Tumor Volume (mm3)']
    FinalTumorVol.append(final_tumor_vol)
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"This data is an outlier: {outliers} in Regimen {regimen}" )
  
```

```python
boxplotFour = plt.boxplot(FinalTumorVol, labels = fourRegimen)
plt.title("Four Regimen Boxplot for Outliers")
plt.xlabel("Regimen")
plt.ylabel("Tumor Volume in mm3")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
CapTable = merge_df.loc[merge_df["Drug Regimen"] == "Capomulin"]
CapMouse = CapTable.loc[CapTable["Mouse ID"] == "s185"]
plt.title("Tumor Volume Over Timepoint for Mouse ID s185 for Capomulin")
plt.xlabel("Timepoint ")
plt.ylabel("Tumor Volume in mm3")
plt.plot(CapMouse['Timepoint'], CapMouse["Tumor Volume (mm3)"])
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
CapTableGroup = CapTable.groupby("Mouse ID").mean()
plt.title("Mouse Weight Vesus Tumor Volume for Mice for Capomulin")
plt.xlabel("Mouse Weight in g ")
plt.ylabel("Tumor Volume in mm3")
plt.scatter(CapTableGroup['Weight (g)'],CapTableGroup["Tumor Volume (mm3)"] )
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(CapTableGroup['Weight (g)'],CapTableGroup["Tumor Volume (mm3)"])
correlation[0]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(CapTableGroup['Weight (g)'],CapTableGroup["Tumor Volume (mm3)"])
regress_values = CapTableGroup['Weight (g)'] * slope + intercept
plt.scatter(CapTableGroup['Weight (g)'],CapTableGroup["Tumor Volume (mm3)"] )
plt.plot(CapTableGroup['Weight (g)'],regress_values,"r-")
plt.title("Mouse Weight Vesus Tumor Volume for Mice for Capomulin")
plt.xlabel("Mouse Weight in g ")
plt.ylabel("Tumor Volume in mm3")
```
